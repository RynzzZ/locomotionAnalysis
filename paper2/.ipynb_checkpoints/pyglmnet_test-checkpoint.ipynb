# pyglmnet test
Test `pyglmnet` to see if faster/better than `glmnet` matlab package...

```python
import matplotlib.pyplot as plt
from scipy.io import loadmat
from pyglmnet import GLM, GLMCV
import numpy as np
import time
import os

%matplotlib notebook

# load data
data = loadmat(os.path.join('glm', 'test.mat'))
X_original, y = data['X'], data['y'][:,0]
```

```python
# glm (no cross validation, no regularization)

# temp
X = X_original[:,:]
X = (X-np.mean(X,0)) / np.std(X,0)

# fit glm and generate predictions
t0 = time.time()
glm = GLM(distr='softplus', alpha=0, reg_lambda=0, max_iter=10000, learning_rate=1e-0,
          score_metric='pseudo_R2', fit_intercept=True, verbose=True)
glm.fit(X, y)
print('finished in {:.1f} minutes'.format((time.time()-t0)/60))
yhat = glm.predict(X)

# score the model on test data
pseudo_R2 = glm.score(X, y)
print('pseudo R^2: {:.2f}'.format(pseudo_R2))

# plot yhat
plt.figure()
plt.plot(yhat);
plt.xlim((0,200*60))
```

```python
# glm (cross validation, regularization)

# temp
X = X_original[:,:]
X = (X-np.mean(X,0)) / np.std(X,0)

# fit glm and generate predictions
lambdas = np.logspace(np.log(1e-1), np.log(1e-8), 50, base=np.exp(1))
t0 = time.time()
glm = GLMCV(distr='softplus', cv=3, alpha=0, reg_lambda=lambdas, max_iter=10000, learning_rate=1e-0,
          score_metric='pseudo_R2', fit_intercept=True, verbose=True)
glm.fit(X, y)
print('finished in {:.1f} minutes'.format((time.time()-t0)/60))
yhat = glm.predict(X)

# score the model on test data
pseudo_R2 = glm.score(X, y)
print('pseudo R^2: {:.2f}'.format(pseudo_R2))

# plot yhat
plt.figure()
plt.plot(yhat);
plt.xlim((0,200*60))
```

```python
plt.figure()
plt.plot(glm.reg_lambda, glm.scores_)
opt_score = glm.scores_[int(np.where(glm.reg_lambda==glm.reg_lambda_opt_)[0])]
plt.scatter(glm.reg_lambda_opt_, opt_score, c='red')
plt.xscale('log')
```

# try their test code...

```python
import numpy as np
import scipy.sparse as sps

import matplotlib.pyplot as plt
from pyglmnet import GLM, simulate_glm

n_samples, n_features = 1000, 100
distr = 'poisson'

# sample a sparse model
np.random.seed(42)
beta0 = np.random.rand()
beta = sps.random(1, n_features, density=0.2).toarray()[0]

# simulate data
Xtrain = np.random.normal(0.0, 1.0, [n_samples, n_features])
ytrain = simulate_glm('poisson', beta0, beta, Xtrain)
Xtest = np.random.normal(0.0, 1.0, [n_samples, n_features])
ytest = simulate_glm('poisson', beta0, beta, Xtest)

# create an instance of the GLM class
glm = GLM(distr='poisson', score_metric='pseudo_R2', reg_lambda=0.01)

# fit the model on the training data
glm.fit(Xtrain, ytrain)

# predict using fitted model on the test data
yhat = glm.predict(Xtest)

# score the model on test data
pseudo_R2 = glm.score(Xtest, ytest)
print('Pseudo R^2 is %.3f' % pseudo_R2)

# plot the true coefficients and the estimated ones
plt.stem(beta, markerfmt='r.', label='True coefficients')
plt.stem(glm.beta_, markerfmt='b.', label='Estimated coefficients')
plt.ylabel(r'$\beta$')
plt.legend(loc='upper right')

# plot the true vs predicted label
plt.figure()
plt.plot(ytest, yhat, '.')
plt.xlabel('True labels')
plt.ylabel('Predicted labels')
plt.plot([0, ytest.max()], [0, ytest.max()], 'r--')
plt.show()
```
